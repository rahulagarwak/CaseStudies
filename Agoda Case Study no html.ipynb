```python
import pandas as pd
df = pd.read_excel('Case_Study_Urgency_Message_Data.xlsx', sheet_name=None)  # reading the excel file
mdf = pd.concat(df,axis=0, ignore_index=True)  # merging the sheets
print(mdf)
mdf.to_csv('output.csv')
```

```python
#Summary Statistics for ADR_USD and star_rating columns
df = pd.DataFrame(mdf,columns=['ADR_USD','star_rating','accommodation_type_name','chain_hotel'])
print(df.describe())
```

```python
#Covariance bnetween ADR_USD and star_rating columns
print(df.cov())
```

```python
import plotly.express as px
fig = px.scatter(df, x='star_rating', y='ADR_USD',color='ADR_USD',facet_col='chain_hotel') #facet_col='accommodation_type_name'
fig.show()
```

```python
#chain vs non chain ratings trend
grouped_data = pd.DataFrame(df.groupby(['accommodation_type_name','chain_hotel'], as_index=False)['star_rating'].mean().sort_values('star_rating'))
fig = px.scatter(grouped_data, x='star_rating', y='accommodation_type_name',facet_row = 'chain_hotel')
fig.show()
```

```python
#chain vs non-chain count
count_data = df.groupby(['chain_hotel'])['chain_hotel'].count()
print(count_data)
```

```python
#Daywise bookings
mdf['Day_number'] = mdf.checkin_date.dt.dayofweek
mdf['Day_name'] = mdf.checkin_date.dt.day_name()
daywise_bookings = mdf.groupby(['Day_name','Day_number'])['checkin_date'].count().reset_index()
daywise_bookings = daywise_bookings.sort_values('Day_number')
daywise_bookings
```

```python
#No of days of booking distribution
fig = px.bar(daywise_bookings, x='Day_name', y='checkin_date',color='Day_number')
fig.show()
```
